```
from matplotlib import pyplot as plt
x1 = [1,5,4]
y1 =[23,45,89]
x2 = [3,7,3]
y2 = [5,8,9] 
plt.plot(x1,y1)
plt.plot(x2,y2)
plt.show()
```

```
from matplotlib import pyplot as plt
from matplotlib import style

style.use('ggplot')
x1 = [1,5,4]
y1 =[23,45,89]
x2 = [3,7,3]
y2 = [5,8,9] 
plt.plot(x1,y1)
plt.plot(x2,y2)
plt.show()
```

```
from matplotlib import pyplot as plt
from matplotlib import style

style.use('ggplot')
x1 = [1,5,4]
y1 =[23,45,89]
x2 = [3,7,3]
y2 = [5,8,9] 
plt.plot(x1,y1, label="first")
plt.plot(x2,y2, label ='second')
plt.legend()
plt.show()
```
